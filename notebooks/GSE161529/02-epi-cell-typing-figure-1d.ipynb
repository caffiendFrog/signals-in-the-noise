## 0. Initializations

```python
# -- imports --
from signals_in_the_noise.preprocessing.gse161529 import GSE161529
from signals_in_the_noise.utilities.visualization import get_figure_axes
```

```python
# -- dataset --
gse = GSE161529()
all_real, all_noise = gse.get_combined_epithilial_dataset(real_pca_kwargs={}, noise_pca_kwargs={}, hvg_post_stromal=True)
```

```python
## -- additionally filter to pre & post --
all_real_pre = all_real[all_real.obs['hormonal_status'] == 'Pre'].copy()
all_real_post = all_real[all_real.obs['hormonal_status'] == 'Post'].copy()

all_noise_pre = all_noise[all_noise.obs['hormonal_status'] == 'Pre'].copy()
all_noise_post = all_noise[all_noise.obs['hormonal_status'] == 'Post'].copy()
```

## 1. "Real" cells

```python
# _, axes = get_figure_axes(2, num_cols=2)
gse.visualize_tsne(all_real_pre, color=['hormonal_status'], plot_kwargs={'title': 'Pre-menopause'})
gse.visualize_tsne(all_real_post, color=['hormonal_status'], plot_kwargs={'title': 'Post-menopause'})
```

```python
gse.visualize_tsne(all_real, color=['hormonal_status'])
```

## 2. "Noise" cells

```python
gse.visualize_tsne(all_noise_pre, color=['hormonal_status'], plot_kwargs={'title': 'Pre-menopause'})
gse.visualize_tsne(all_noise_post, color=['hormonal_status'], plot_kwargs={'title': 'Post-menopause'})
```
